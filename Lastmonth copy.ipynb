```python
import Portfolio
from Narrowing import knn 
from Narrowing import price
from DataColection import TickerPuller
import time

from Narrowing import adr
from Narrowing import volume

import imp
import mypytable
from Predicting import HiFinder 
import math
```

```python
user = Portfolio.portfolio()

user.cash=100


```

```python
l = 30

while l > 5:
    # try:
        print("iteration")
        print(l)

        recomended,day = knn.knn_past(days_back=l)
        
        print("out of knn")
        inbudget,day = price.inbudget(recomended,user.cash,5,day=day)
        # from Narrowing import adr
        # from Narrowing import volume

        # print(len(inbudget))
        # print(len(day))
        good_volume,day = volume.remove_by_volume(inbudget,250000,250000,days=day)
        # print(len(good_volume))
        # print(len(day))


        stocks,day = adr.best_daily_range(good_volume,5,days=day)
        # print(len(stocks))
        # print(len(day))

        # print(stocks)

        # print(len(day))
        # print(day)
        stonks = []
        for val in stocks:
            # print(val)
            stonks.append(val.replace(".csv",""))
        for i ,stock in enumerate(stonks):
            TickerPuller.thisday(stock,int(day[i]))
            time.sleep(1.1)
        # print(day)
        # import imp
        # import mypytable
        # from Predicting import HiFinder 
        # import math
        # imp.reload(HiFinder)
        # imp.reload(Portfolio)
        bought = []
        money = user.cash
        amount_per_stock = money/5
        for name in stocks:
            mt = mypytable.MyPyTable()
            mt.load_from_file("Data/todays/"+name)
            
            past = HiFinder.StanderdStock(name)

            avg = past.Low_mean_std()[0]
            std = past.Low_mean_std()[1]
            # print(avg,std)
            mockrun = mt.get_column('o')
            start = mockrun[0]
            mockrun.pop(0)
            for i,val in enumerate(mockrun):
                if (start - (avg - std)) >=val:
                    
                    user.liquid[name] = math.floor(amount_per_stock/val)
                    user.price[name] = val
                    user.cash = user.cash - (val * math.floor(amount_per_stock/val))

                    
                    bought.append(name)
                    
                    break
        
        print("Cash after buying:",end=" ")
        print(user.cash)

        print("keys:",end="")
        print(user.liquid.keys())

        stonks = []
        for val in user.liquid.keys():
            # print(val)
            stonks.append(val.replace(".csv",""))
        for i ,stock in enumerate(stonks):
            TickerPuller.thisday(stock,int(day[0] + 86400 ))
            time.sleep(1.1)
        print("Amount of stock:",end=" ")
        print(len(stocks))
        for name in bought:
            sellifbellow = False
            sellnow = False
            time_to_sell = 0
            trail_ammount =30
            mt = mypytable.MyPyTable()
            mt.load_from_file("Data/todays/"+name)

            past = HiFinder.StanderdStock(name)

            avg = past.Hi_mean_std()[0]
            std = past.Hi_mean_std()[1]
            # print(avg,std)
            mockrun = mt.get_column('o')
            start = mockrun[0]
            trail = start
            if user.price[name]<start:
                trail_ammount = 20
            mockrun.pop(0)
            
            for i,val in enumerate(mockrun):
                if (start + (avg + std)) <=val:
                    user.cash += user.liquid[name] * val
                    user.profits +=  (user.liquid[name]*val) - (user.price[name]*user.liquid[name])
                    break
                    
                # if (start + (avg )) <=val:
                #     trail_ammount = 10
                    

                # if val > trail:
                #     trail = val
                #     time_to_sell =0
                # else:
                #     time_to_sell +=1
                
                # if time_to_sell>=trail_ammount:
                #     print("sold from trail")
                #     user.cash += user.liquid[name] * val
                #     user.profits +=  (user.liquid[name]*val) - (user.price[name]*user.liquid[name])
                #     break



                

                if i> len (mockrun)*.9 :
                    user.cash += user.liquid[name] * val
                    user.profits +=  (user.liquid[name]*val) - (user.price[name]*user.liquid[name])

                    # user.price.pop(name)
                    print("Here")
                    break
        user.liquid.clear()        
        l-=1    
        print("profits:",end=" ")
        print(user.profits)
        print("cash:",end=" ")
        print(user.cash)
```
