```python
import os
import pandas as pd
import numpy as np
import mypytable as mypy
from sklearn.model_selection import train_test_split
```

```python
stockfiles= os.listdir("/home/cbarker4/Documents/DataScience/StockTrader/Data")
print(stockfiles)
```

```python
X = []
Y=[]
for val in stockfiles:


    mt = mypy.MyPyTable()
    mt.load_from_file("/home/cbarker4/Documents/DataScience/StockTrader/Data/"+stockfiles[0])
    mt.drop_column('v')
    mt.drop_column('t')
    i = 100
    while i <  + len(mt.data)-1:
        table = mt.create_sub_table(i-100,i)

        table.drop_column('v')
        table.drop_column('t')
        X.append(table.data)
        Y.append(mt.get_row(i+1))
        i+=1
```

```python
X = np.array(X)

Y = np.array(Y)
print(Y.shape)
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=.25 ,random_state=0)
# print('X_train shape: ', X_train.shape)
```
